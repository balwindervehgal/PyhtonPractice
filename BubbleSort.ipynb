# BubbleSort

The bubble sort makes multiple passes through a list. It compares adjacent items and exchanges those that are out of order. Each pass through the list places the next largest value in its proper place. In essence, each item “bubbles” up to the location where it belongs.

Below diagram shows the first pass of a bubble sort. The shaded items are being compared to see if they are out of order. If there are n items in the list, then there are n−1 pairs of items that need to be compared on the first pass. It is important to note that once the largest value in the list is part of a pair, it will continually be moved along until the pass is complete.



```python
def bubbleSort(arr):
    
    str_len = len(arr)
    
    # looping through all elements
    for index in range(str_len):
        
        #compare each element with rest of the elements and shift accordingly
        #note the range as last indexed elements alreasy in place
        
        for nested_index in range(str_len - index -1):
        
            #check if need shift
            if arr[nested_index] > arr[nested_index+1]:
            
                #swap elements
                arr[nested_index], arr[nested_index+1] = arr[nested_index+1], arr[nested_index]
    
    #return sorted array/list
    return arr
            
```

```python
print (bubbleSort([54,26,93,17,77,31,44,55,20]))
```
